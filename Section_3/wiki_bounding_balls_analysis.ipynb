```python
import pandas as pd
import numpy as np
import math
from scipy import stats
import matplotlib.pyplot as plt
```

```python
sentences = pd.read_csv("30K_random_words_sentences.csv")
sentences = sentences.groupby("Unnamed: 0").count().reset_index()
print("Total number of words with 10 examples:", len(sentences[sentences['0'] >=10]))
```

```python
data = pd.read_csv("30k_random_words_bounding_balls.csv")
print("Total number of words bounding ball results:", len(data))
```

```python
data['Word Frequency (logged)'] = data['freq'].apply(lambda x: math.log(x))
data['Radius of Bounding Ball'] = data['r2'].apply(lambda x: math.sqrt(x))
print(stats.pearsonr(x = data['Radius of Bounding Ball'].values, y = data['Word Frequency (logged)'].values))
```

```python
fig = plt.figure()
plt.scatter(x = data['Word Frequency (logged)'].values, y=data['Radius of Bounding Ball'].values, s=0.1, rasterized=True)
plt.tick_params(axis='both', which='major', labelsize=15)
plt.xlabel('Word Frequency (logged)', fontsize=18)
plt.ylabel('Radius of Bounding Ball', fontsize=18)
plt.savefig('radii_word_freq.pdf', bbox_inches='tight', dpi=600)
```
