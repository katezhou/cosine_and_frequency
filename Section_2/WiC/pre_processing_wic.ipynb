```python
def pre_process_wic_data(filename):
    data = pd.read_csv("../WiC_data/" + filename + "/" + filename + ".data.txt", sep='\t',header=None)
    data.columns = ['keyword', 'POS', 'indices', 'sentence1', 'sentence2']

    data['index1'] = data['indices'].apply(lambda x: x.split('-')[0])
    data['index2'] = data['indices'].apply(lambda x: x.split('-')[1])

    keyword1s = []
    keyword2s = []

    for n, row in data.iterrows():
        keyword1s.append(row['sentence1'].split(" ")[int(row['index1'])])
        keyword2s.append(row['sentence2'].split(" ")[int(row['index2'])])
    data['keyword1'] = keyword1s
    data['keyword2'] = keyword2s
    data['index'] = data.index

    data_a = data[['index', 'keyword1', 'POS', 'sentence1', 'index1']]
    data_b = data[['index', 'keyword2', 'POS', 'sentence2', 'index2']]
    data_a.columns = ['index', 'keyword', 'POS', 'sentence', 'word_index']
    data_b.columns = ['index', 'keyword', 'POS', 'sentence', 'word_index']

    data = pd.concat([data_a, data_b], axis=0)
    data.to_csv("sentences/" + filename + "_sentences.csv", index=False)
```

```python
pre_process_wic_data("dev")
pre_process_wic_data("train")
```
